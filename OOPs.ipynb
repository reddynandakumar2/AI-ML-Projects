```python
class MLAlgorithm:
    def __init__(self,algorithm,accuracy):
        self.algorithm=algorithm
        self.accuracy=accuracy
        self.evaluate()
    def evaluate(self):
        print("Algorithm: ",self.algorithm)
        print("Accuracy: ",self.accuracy)
s1=MLAlgorithm("XGBoost","92%")
```

```python
class ModelTrainer:
    def __init__(self,model_name,epochs):
        self.model_name=model_name
        self.epochs=epochs
        s1.train()
        s1.show_epochs()
    def train(self):
        print(f"{self.model_name} is successfully Trained")
    def show_epochs(self):
        print(f"{self.epochs} epochs are used while training")
s1=ModelTrainer("CNN",100)
```

```python
class Metrics:
    def __init__(self,accuracy,precision,recall):
        self.accuracy=accuracy
        self.precision=precision
        self.recall=recall
    def display_metrics(self):
        print("Accuracy: ",self.accuracy)
        print("Precision: ",self.precision)
        print("Recall: ",self.recall)
s1=Metrics(90,92,94)
s1.display_metrics()
```

```python
class anyclass:
    def __init__(self,input_size,Hidden_size,output_size):
        self.input_size=input_size
        self.Hidden_size=Hidden_size
        self.output_size=output_size
    def forward(self):
        print("Input_Size:",self.input_size)
    def backward(self):
        print("Hidden_Size:",self.Hidden_size)
    def output(self):
        print("Output_Size:",self.output_size)
s1=anyclass(32,64,4)
s1.forward()
s1.backward()
s1.output()
```
