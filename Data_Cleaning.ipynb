```python
# Handling Missing Values
import pandas as pd
df=pd.read_excel("C:/Users/Lenovo/Downloads/Handling_Missing_Values_Practice.xlsx")
df=df.drop(columns=['Student_ID'])
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Marks']=df['Marks'].fillna(df['Marks'].median())
df['City']=df['City'].fillna(df['City'].mode()[0])
df['Study_Hours']=df['Study_Hours'].fillna(df['Study_Hours'].mean())
print(df)
```

```python
# Categorical Encoding
import pandas as pd
df=pd.read_excel("C:/Users/Lenovo/Downloads/Categorical_Encoding_Practice.xlsx")
df=df.drop(columns=['Student_ID'])
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
encoder=LabelEncoder()
encoder2=OneHotEncoder(sparse_output=False,
    handle_unknown='ignore')
df['Gender']=encoder.fit_transform(df['Gender'])
encoded=encoder2.fit_transform(df[['City','Education_Level','Course']])
encoded_df = pd.DataFrame(
    encoded,
    columns=encoder2.get_feature_names_out(
        ['City','Education_Level','Course']
    )
)
df = pd.concat(
    [df, encoded_df],
    axis=1
)
df=df.drop(columns=['City','Education_Level','Course'])
print(df)


```

```python
# Feature Scaling
import pandas as pd
from sklearn.preprocessing import StandardScaler
df=pd.read_excel("C:/Users/Lenovo/Downloads/Feature_Scaling_Practice.xlsx")
df=df.drop(columns=['Employee_ID'])
scaler=StandardScaler()
df['Age']=scaler.fit_transform(df[['Age']])
df['Salary']=scaler.fit_transform(df[['Salary']])
df['Experience_Years']=scaler.fit_transform(df[['Experience_Years']])
df['Distance_From_Office']=scaler.fit_transform(df[['Distance_From_Office']])
print(df)
corr_matrix = df.corr(numeric_only=True)

print(corr_matrix)
```
